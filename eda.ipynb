# Arabert

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("elkhatib/final_try",use_auth_token=True,model_max_length=512)
```

```python
def preprocess_function(examples):
    return tokenizer(examples["text"],truncation=True,padding=True)
```

```python
tokenized_ds = ds.map(preprocess_function,batched=True)
```

```python
tokenized_ds = tokenized_ds["train"].train_test_split(0.2)
```

```python
from transformers import DataCollatorWithPadding

data_collator = DataCollatorWithPadding(tokenizer=tokenizer)
```

```python
import evaluate
accuracy = evaluate.load("accuracy")
```

```python
import numpy as np 

def compute_metrics(eval_pred):
    predictions,labels = eval_pred
    predictions = np.argmax(predictions,axis=1)
    return accuracy.compute(predictions=predictions,references=labels)
```

```python
id2label = {0:"Spam",1:"Ham"}
label2id = {"Spam":0,"Ham":1}
```

```python
from transformers import AutoModelForSequenceClassification, TrainingArguments, Trainer
model = AutoModelForSeq2SeqLM.from_pretrained("elkhatib/final_try",use_auth_token=True,num_labels=2,
                                              id2label=id2label,label2id=label2id)
```

```python
training_args = TrainingArguments(
    output_dir="my_spam_filter_model",
    learning_rate=2e-5,
    per_device_train_batch_size=4,
    per_device_eval_batch_size=4,
    num_train_epochs=2,
    weight_decay=0.01,
    save_total_limit=2,
    # push_to_hub=True
    eval_accumulation_steps=16,
    load_best_model_at_end=True,
    save_strategy="epoch",
    evaluation_strategy="epoch",
    report_to="none"
    
)
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset =tokenized_ds["train"],
    eval_dataset=tokenized_ds["test"],
    tokenizer=tokenizer,
    data_collator=data_collator,
    compute_metrics=compute_metrics,
    
)
trainer.train()
```
